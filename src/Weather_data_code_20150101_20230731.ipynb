```python
import pandas as pd
```

```python
df_air_temperature = pd.read_csv("../data/freiburg weather data before 23-08-05/temperature_produkt_tu_termin_18740101_20231231_01443.txt",
                                 delimiter=";", usecols=["MESS_DATUM", "RF_TER", "TT_TER"])
df_air_temperature = df_air_temperature.rename(columns={
    "MESS_DATUM": "Measurement_Date",
    "RF_TER": "Relative_Humidity",
    "TT_TER": "Air_Temperature"
})

# Convert 'Measurement_Date' to datetime
df_air_temperature['Measurement_Date'] = pd.to_datetime(df_air_temperature['Measurement_Date'], format='%Y%m%d%H')

# Set the datetime column as the index
df_air_temperature.set_index('Measurement_Date', inplace=True)

# Resample to daily, aggregating with mean for temperature and humidity
df_air_temperature_daily = df_air_temperature.resample('D').agg({
    'Air_Temperature': 'mean',   # Average temperature
    'Relative_Humidity': 'mean'  # Average humidity
})
# Filter data to keep entries from 2015 onward
df_air_temperature_daily = df_air_temperature_daily.loc['2015-01-01': '2023-07-31']
```

```python
from IPython.display import display
display(df_air_temperature_daily.head(5))
display(df_air_temperature_daily.tail(5))
```

```python
df_cloudiness = pd.read_csv("../data/freiburg weather data before 23-08-05/cloudiness_produkt_n_termin_18740101_20231231_01443.txt", delimiter=";", usecols=["MESS_DATUM","N_TER"])
df_cloudiness = df_cloudiness.rename(columns={
    "MESS_DATUM": "Measurement_Date",
    "N_TER": "Cloud_Cover"
})

df_cloudiness['Measurement_Date'] = pd.to_datetime(df_cloudiness['Measurement_Date'], format='%Y%m%d%H')

df_cloudiness.set_index('Measurement_Date', inplace=True)

df_cloudiness_daily = df_cloudiness.resample('D').agg({
    'Cloud_Cover': 'mean',   
})
# Filter data to keep entries from 2015 onward
df_cloudiness_daily = df_cloudiness_daily.loc['2015-01-01': '2023-07-31']
```

```python
display(df_cloudiness_daily.head(5))
display(df_cloudiness_daily.tail(5))
```

```python
df_visibility = pd.read_csv("../data/freiburg weather data before 23-08-05/visibility_produkt_vk_termin_19490501_20231231_01443.txt", delimiter=";", usecols=["MESS_DATUM","VK_TER"])

df_visibility = df_visibility.rename(columns={
    "MESS_DATUM": "Measurement_Date",
    "VK_TER": "Visibility"
})
# Convert 'Measurement_Date' to datetime
df_visibility['Measurement_Date'] = pd.to_datetime(df_visibility['Measurement_Date'], format='%Y%m%d%H')

# Set the datetime column as the index
df_visibility.set_index('Measurement_Date', inplace=True)

# Resample to daily, aggregating with mean for visibility
df_visibility_daily = df_visibility.resample('D').agg({
    'Visibility': 'mean'  # Average visibility for the day
})
df_visibility_daily = df_visibility_daily.loc['2015-01-01': '2023-07-31']
```

```python
display(df_visibility_daily.head(5))
display(df_visibility_daily.tail(5))
```

```python
df_precipitation = pd.read_csv(
    "../data/freiburg weather data before 23-08-05/Precipitation_produkt_nieder_tag_18740101_20231231_01443.txt", delimiter=";", usecols=["MESS_DATUM","  RS", "SH_TAG"])

df_precipitation = df_precipitation.rename(columns={
    "MESS_DATUM": "Measurement_Date",
    "  RS": "Daily_Precipitation_Height",
    "SH_TAG": "Snow_Height_Daily_Value"
})

df_precipitation['Measurement_Date'] = pd.to_datetime(df_precipitation['Measurement_Date'], format='%Y%m%d')

df_precipitation.set_index('Measurement_Date', inplace=True)

df_precipitation_daily = df_precipitation

df_precipitation_daily = df_precipitation_daily.loc['2015-01-01': '2023-07-31']
```

```python
df_solar = pd.read_csv(
    "../data/freiburg weather data before 23-08-05/Solar_irradiance_Freiburg_Dach_2015-01-01_2023-07-31.csv", delimiter=",", usecols=["Beginn-Lokalzeit","Globalstrahlung (W/m²)"])

df_solar = df_solar.rename(columns={
    "Beginn-Lokalzeit": "Measurement_Date",
    "Globalstrahlung (W/m²)": "Global_Radiation"
})

df_solar['Measurement_Date'] = pd.to_datetime(df_solar['Measurement_Date'], format='%Y-%m-%d %H:%M:%S')  # Adjust the format if needed

df_solar.set_index('Measurement_Date', inplace=True)

df_solar_daily = df_solar.resample('D').sum()
```

```python
display(df_solar_daily.head(5))
display(df_solar_daily.tail(5))
```

```python
df_wind = pd.read_csv(
    "../data/freiburg weather data before 23-08-05/wind_produkt_fk_termin_18740101_20231231_01443.txt", delimiter=";", usecols=["MESS_DATUM","DK_TER", "FK_TER"])

df_wind = df_wind.rename(columns={
    "MESS_DATUM": "Measurement_Date",
    "DK_TER": "Wind_Direction",
    "FK_TER": "Wind_Strength"
})

df_wind['Measurement_Date'] = pd.to_datetime(df_wind['Measurement_Date'], format='%Y%m%d%H')

df_wind.set_index('Measurement_Date', inplace=True)

df_wind_daily = df_wind.resample('D').agg({
    'Wind_Direction': 'mean',   
    'Wind_Strength': 'mean'  
})
df_wind_daily = df_wind_daily.loc['2015-01-01': '2023-07-31']
```

```python
display(df_wind_daily.head(5))
display(df_wind_daily.tail(5))
```

```python
df_air_pressure = pd.read_csv(
    "../data/freiburg weather data before 23-08-05/air_pressure_produkt_pp_termin_18740101_20231231_01443.txt", delimiter=";", usecols=["MESS_DATUM","PP_TER"])

df_air_pressure = df_air_pressure.rename(columns={
    "MESS_DATUM": "Measurement_Date",
    "PP_TER": "Air_Pressure_at_Station_Height"
})
df_air_pressure['Measurement_Date'] = pd.to_datetime(df_air_pressure['Measurement_Date'], format='%Y%m%d%H')

df_air_pressure.set_index('Measurement_Date', inplace=True)
df_air_pressure
df_air_pressure_daily = df_air_pressure.resample('D').agg({
    'Air_Pressure_at_Station_Height': 'mean',
})
df_air_pressure_daily = df_air_pressure_daily.loc['2015-01-01': '2023-07-31']
```

```python
df_vapour_pressure_and_rel_humidity = pd.read_csv(
    "../data/freiburg weather data before 23-08-05/vapour_pressure_relative_humidity_produkt_tf_termin_18740101_20231231_01443.txt", delimiter=";", usecols=["MESS_DATUM", "RF_TER", "TF_TER", "VP_TER"])

df_vapour_pressure_and_rel_humidity = df_vapour_pressure_and_rel_humidity.rename(columns={
    "MESS_DATUM": "Measurement_Date",
    "RF_TER": "Relative_Humidity",
    "TF_TER": "Dew_Point_Temperature",
    "VP_TER": "Vapor_Pressure"
})

df_vapour_pressure_and_rel_humidity['Measurement_Date'] = pd.to_datetime(
    df_vapour_pressure_and_rel_humidity["Measurement_Date"], format="%Y%m%d%H")

df_vapour_pressure_and_rel_humidity.set_index('Measurement_Date', inplace=True)

df_vapour_pressure_and_rel_humidity_daily = df_vapour_pressure_and_rel_humidity.resample('D').agg({
    "Vapor_Pressure": "mean",
    "Dew_Point_Temperature": 'mean'}
)
df_vapour_pressure_and_rel_humidity_daily = df_vapour_pressure_and_rel_humidity_daily.loc['2015-01-01': '2023-07-31']
```

```python
display(df_vapour_pressure_and_rel_humidity_daily.head(5))
display(df_vapour_pressure_and_rel_humidity_daily.tail(5))
```

```python
dataframes = [df_air_temperature_daily, df_visibility_daily, df_air_pressure_daily, df_cloudiness_daily, df_precipitation_daily, df_solar_daily, df_vapour_pressure_and_rel_humidity_daily, df_wind_daily]

# Concatenate all DataFrames along columns (axis=1), aligning by the index
df_combined = pd.concat(dataframes, axis=1)

#round all the entry to 3 decimal places
df_combined = df_combined.round(2)
# Display the concatenated DataFrame
df_combined
```

```python
#write a csv file
df_combined.to_csv("../data/freiburg_weather_150101_230805.csv", sep=";")
```
